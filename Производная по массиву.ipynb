```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def diff(x,y):
    N = len(x)
    result = np.zeros(N)
    i=0   
    while i<N-1:
        result[i] = (y[i+1]-y[i])/(x[i+1]-x[i])
        i+=1
    else:
        result[i] = result[i-1]
    return result
```

```python
N=100
dx = 2*np.pi/(N-1)
x = np.arange(0, 2*np.pi+dx, dx)
y = np.sin(x)
    
y1 = diff(x,y)
plt.plot(x,y1)
plt.show()
```
